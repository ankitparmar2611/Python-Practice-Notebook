```python
# Trying to use a Python keyword as an identifier gives SyntaxError.
# But we can check with keyword module
# Do not ever use the variable which is same as keyword names as you see in this cells output

import keyword
print("Python keywords are:")
print(keyword.kwlist)

# Example tricky check:
print("Is 'lambda' a keyword?", keyword.iskeyword("lambda"))
print("Is 'Lambda' a keyword?", keyword.iskeyword("Lambda"))  # case-sensitive
```

```python
# Here we have craeted a fucntion outer() and created a for loop in range 0 to 5

# Below is the Multiline comment we have created using """ """.  Text written inside this is a multiline comment""" 

"""
Here 

First line executes:
Start outer

Loop Starts:
for i in range(5):

0, 1, 2, 3, 4
So loop runs 5 times.

i = 0

0 % 2 = 0 (0/2)= 0, so it is even.
Condition true → even

Output:
even inside loop 0

1 % 2 = 1
Condition false → odd

Output:
Odd inside loop 1

2 % 2 = 1
Condition True → Even

Output:
even inside loop 2

etc..... till  4 % 2

"""
def outer():
    print("Start outer")
    for i in range(5):
        if i % 2 == 0:
            print("even inside loop", i)
        else:
            print("Odd inside loop", i)
    print("End Outer")
    
outer()
```
