```python
# 5️ Function with variable-length arguments (*args)

# Example 1
def multiply(*numbers):
    """
    This function multiplies all numbers passed to it.
    *numbers means the function can accept any number of arguments.
    All the arguments are stored inside a tuple called 'numbers'.
    """
    
    # Print the tuple of numbers received by the function
    print(numbers)

    # We start result with 1 because multiplication identity is 1
    # If we started with 0, the result would always become 0
    result = 1  
    
    # Loop through each number in the tuple
    for num in numbers:
        
        # Print the current number being processed
        print(num)

        # Multiply the current number with result
        # Example calculation:
        # First iteration: result = 1 * 2 = 2
        # Second iteration: result = 2 * 3 = 6
        # Third iteration: result = 6 * 4 = 24
        result *= num
        
    # After the loop finishes, return the final multiplication result    
    return result
    
# Calling the function with numbers
print("Product:", multiply(2,3,4))
  
# Example 2

def mul(*number):
    print(number)
    result1 = 1
    for num in number:
        result1 *= num
    return result1
print("Result :", mul(2,4,6))   
    
```

```python
# ----------------------------------------
# Using 'global' keyword to modify global variables
# ----------------------------------------

# Example 1
# A global variable 'count' is created outside the function
count = 200

def increment():
    global count            # This tells Python we want to use and modify the global variable
    count += 1              # Increase the value of count by 1
    print(f"Inner count = {count}")   # Print the updated value

increment()  # First call → count becomes 201
increment()  # Second call → count becomes 202


# ----------------------------------------

# Example 2
# Global variable 'number' starts with value 300
number = 300

def number_list():
    global number           # Access and modify the global variable
    number += 1             # Increase number by 1 each time function runs
    print(f"Number List : {number}")  # Print the updated value

number_list()   # 301
number_list()   # 302
number_list()   # 303


# ----------------------------------------

# Example 3
# Global variable 'even' starts from 2 (first even number)
even = 2

def even_number():
    global even             # Use the global variable 'even'
    print(f"Even number : {even}")  # Print current even number
    even += 2               # Increase by 2 so the next even number is stored

even_number()   # Prints 2
even_number()   # Prints 4
even_number()   # Prints 6


# ----------------------------------------

# Example 4
# Global variable 'odd_number' starts from 1 (first odd number)
odd_number = 1 

def odd():
    global odd_number       # Access the global variable
    print(f"Odd number : {odd_number}")  # Print current odd number
    odd_number += 2         # Increase by 2 to get the next odd number

odd()   # Prints 1
odd()   # Prints 3
odd()   # Prints 5
odd()   # Prints 7
odd()   # Prints 9
```

```python
# Generator function to print even numbers between 1 and n

def number_gen(n):                     # Define a generator function that accepts a limit 'n'
    for i in range(1, n+1):            # Loop through numbers from 1 to n
        if i % 2 == 0:                 # Check if the number is even
            yield i                    # If even, send the number to the caller and pause


# Loop through the generator and print each value
for num in number_gen(20):             # Call the generator with n = 20
    print(num)                         # Print each even number produced by the generator
```

```python
# Generator function to produce odd numbers from 1 to n

def odd_generator(n):                # Define a generator function with limit n
    for i in range(1, n+1):          # Loop through numbers from 1 to n
        if i % 2 != 0:               # Check if the number is odd
            yield i                  # If odd, return the number and pause the function


# Loop through the generator and print each odd number
for num in odd_generator(20):        # Call the generator with n = 20
    print(num)                       # Print each odd number produced
```

```python
# Generator function that yields items from a list one by one

def generator(number):              # Define a generator function that takes a list as input
    for num in number:              # Loop through each element in the list
        yield num                   # Return the current element and pause the function


random_list = [20,22,22,36,54,68]   # A sample list of numbers

gen = generator(random_list)        # Create a generator object from the function

print(gen)                          # Prints the generator object (not the values)

print(next(gen))                    # Gets the first value from the generator → 20

print(list(gen))                    # Convert remaining generator values into a list
                                    # Since 20 is already consumed, it prints the rest
```

```python
# Create an infinite generator that produces numbers starting from 1

def infinite_gen():                 # Define a generator function
    num = 1                         # Start the sequence from 1

    while True:                     # Infinite loop (this will run forever)
        yield num                   # Send the current value of num to the caller
                                    # and pause the function here

        num += 1                    # Increase num by 1 to generate the next number


gen = infinite_gen()                # Create a generator object from the function

# next() asks the generator for the next value each time
print(next(gen))                    # First number → 1
print(next(gen))                    # Second number → 2
print(next(gen))                    # Third number → 3
print(next(gen))                    # Fourth number → 4
print(next(gen))                    # Fifth number → 5
print(next(gen))                    # Sixth number → 6
```

### MAP()

```python
# List of numbers
numbers = [1, 2, 3, 4]

# map() applies a function to every element of the iterable (numbers list)
# Syntax: map(function, iterable)

# lambda x: x*x → takes each number and returns its square
result = map(lambda x: x * x, numbers)

# lambda x: x % 2 == 0 → checks if a number is even
# If remainder is 0 → True (even number)
# If remainder is not 0 → False (odd number)
even = map(lambda x: x % 2 == 0, numbers)

# list() converts the map object (iterator) into a list so we can see the values
# list(result) → squares of numbers
# list(even) → True/False values showing if numbers are even
print(f"{list(result)}, {list(even)}")
```

```python
# Import reduce function from functools module
# reduce() is used to apply a function repeatedly to elements of a list
from functools import reduce

# List of numbers
numbers = [3, 2, 5, 4, 1]

# reduce() will compare numbers two at a time
# lambda x, y: x if x > y else y
# This means:
# if x is greater than y → return x
# otherwise → return y
# So this keeps the larger number each time

max_number = reduce(lambda x, y: x if x > y else y, numbers)

# Print the maximum number from the list
print(max_number)
```

```python
# Import reduce function from functools module
# reduce() is used to combine all values of a list into one result
from functools import reduce

# List of numbers
numbers = [5, 10, 15, 20, 25]

# filter(lambda x: x > 10, numbers)
# Step 1: Keep only numbers greater than 10
# Result after filter → [15, 20, 25]

# map(lambda x: x * 2, ...)
# Step 2: Multiply each remaining number by 2
# Result after map → [30, 40, 50]

# reduce(lambda x, y: x + y, ...)
# Step 3: Add all numbers together step by step
# 30 + 40 = 70
# 70 + 50 = 120

# Final result will be 120
result = reduce(lambda x, y: x + y,
                map(lambda x: x * 2,
                    filter(lambda x: x > 10, numbers)))

# Print the final result
print(result)
```

```python
# Import reduce function from functools module
# reduce() is used to combine all values into a single result
from functools import reduce

# List of numbers
numbers = [10, 25, 30, 15, 40]

# filter(lambda x: x > 20, numbers)
# Step 1: Select numbers greater than 20
# Result after filter → [25, 30, 40]

# map(lambda x: x * 3, ...)
# Step 2: Multiply each selected number by 3
# Result after map → [75, 90, 120]

# reduce(lambda x, y: x + y, ...)
# Step 3: Add all numbers step by step
# 75 + 90 = 165
# 165 + 120 = 285

# Final result will be 285
result = reduce(
    lambda x, y: x + y,
    map(lambda x: x * 3,
        filter(lambda x: x > 20, numbers))
)

# Print the final result
print(result)
```

```python
# Creating dictionaries using zip()

# zip() is used to combine two lists element-by-element.
# dict() converts those paired elements into a dictionary.


# ---------------- Example 1 ----------------

# List of student names
names = ["Ankit", "Priya", "Komal"]

# List of marks of each student
marks = [85, 90, 78]

# zip(names, marks) pairs each name with the corresponding mark
# ('Ankit', 85), ('Priya', 90), ('Komal', 78)

# dict() converts the pairs into a dictionary
student_dic = dict(zip(names, marks))

# Print the student dictionary
print(student_dic)


# ---------------- Example 2 ----------------

# List of user names
user_list = ["Jayesh", "Ramesh", "Mahesh"]

# List of user numbers
user_num = [26, 90, 55]

# zip() pairs each user with their number
# ('Jayesh',26), ('Ramesh',90), ('Mahesh',55)

# Convert pairs into dictionary
user_dic = dict(zip(user_list, user_num))

# Print the user dictionary
print(user_dic)


# ---------------- Example 3 ----------------

# List of product names
product = ["Washing Machine", "Laptop", "Dress", "Bed", "T-shirt"]

# List of product categories
pro_category = ["Electronics", "Electronics", "Clothing", "Home Maker", "Clothing"]

# zip() pairs each product with its category
# ('Washing Machine','Electronics')
# ('Laptop','Electronics')
# ('Dress','Clothing')
# ('Bed','Home Maker')
# ('T-shirt','Clothing')

# Convert pairs into dictionary
pro_cat = dict(zip(product, pro_category))

# Print product dictionary
print(pro_cat)


# ---------------- Example 4 ----------------

# List of employee names
employee_name = ["Ankit", "Karan", "John"]

# List of companies where they work
company_name = ["Google", "esparkbiz", "iflair"]

# zip() pairs each employee with their company
# ('Ankit','Google'), ('Karan','esparkbiz'), ('John','iflair')

# Convert pairs into dictionary
emp_dict = dict(zip(employee_name, company_name))

# Print employee dictionary
print(emp_dict)
```

### Accumulate()

```python
# Import accumulate function from itertools module
from itertools import accumulate

# Create a list of numbers
numbers = [1, 2, 3, 4]

# accumulate() keeps adding numbers one by one
# It returns running totals (cumulative sum)
# Steps internally:
# 1
# 1 + 2 = 3
# 3 + 3 = 6
# 6 + 4 = 10
result = accumulate(numbers)

# accumulate returns an iterator, so we convert it to list to see all values
print(list(result))  # Output: [1, 3, 6, 10]
```

```python
# Import accumulate function from itertools module
from itertools import accumulate

# Import operator module (contains functions like add, mul, sub etc.)
import operator

# Create a list of numbers
number = [1, 2, 3, 4]

# accumulate() normally adds numbers
# But here we pass operator.mul which means multiplication instead of addition

# Step-by-step working:
# 1
# 1 * 2 = 2
# 2 * 3 = 6
# 6 * 4 = 24

# Convert result iterator to list and print it
print(list(accumulate(number, operator.mul)))  # Output: [1, 2, 6, 24]
```

```python
# Creating a class (Blueprint)
class user_detail:

    # Constructor method
    # It runs automatically when an object is created
    def __init__(self, name):
        self.name = name      # storing the name inside the object

    # Instance method
    # A method that works with the object data
    def greet(self):
        # printing a greeting using the object's name
        print(f"Hello {self.name}")


# Creating an object of the class
s1 = user_detail("Ankit")   # "Ankit" is passed to the constructor

# Calling the instance method using the object
s1.greet()   # Output: Hello Ankit
```

```python
students = ["Ankit","Priya","Rahul","Neha","Komal"]
cities = ["Ahmedabad","Delhi","Ahmedabad","Pune","Delhi"]
scores = [[85,90,80],[72,88,95],[90,85,92],[60,70,65],[88,90,80]]

# Step 1: Total scores
total_score = [sum(s) for s in scores]

# Step 2: Combine data
student_data = list(zip(students, cities, total_score))

# Step 3: Group by city
city_dict = {}
for name, city, score in student_data:
    if city not in city_dict:
        city_dict[city] = []
    city_dict[city].append((name, score))

# Step 4: Sort and rank
for city in city_dict:
    # Sort by score (highest first)
    city_dict[city].sort(key=lambda x: x[1], reverse=True)
    
    # Assign rank
    city_dict[city] = [
        (name, score, rank + 1)
        for rank, (name, score) in enumerate(city_dict[city])
    ]

# Final output
print(city_dict)
```

```python
# Student Performance Analyzer

student_data = {}
total_marks = 0
average = 0
high_mark = 0
high_name = ""
lowest_mark = None
low_name = ""
pass_students = []
fail_students = []

# Input loop
while True:
    student_name = input("Enter Name: ")
    student_mark = int(input("Enter Marks: "))
    
    student_data[student_name] = student_mark

    choice = input("Do you want to add another student? Y/N: ")
    if choice.lower() != 'y':
        break

# Processing
for index, (name, mark) in enumerate(student_data.items(), start=1):
    # Total Marks
    total_marks += mark
    
    # Highest and lowest marks
    if mark > high_mark:
        high_mark = mark
        high_name = name
    if lowest_mark is None or mark < lowest_mark:
        lowest_mark = mark
        low_name = name

    # Pass/Fail
    if mark >= 40:
        status = "Pass"
        pass_students.append(name)
    else:
        status = "Fail"
        fail_students.append(name)
        
    print(f"{index}. {name} scored {mark} - Status: {status}")

# Calculate average AFTER the loop
average = total_marks / len(student_data)

# Summary
print(f"\nTotal Marks of class: {total_marks}")
print(f"Average Marks: {average}")
print(f"Highest Marks: {high_mark} by {high_name}")
print(f"Lowest Marks: {lowest_mark} by {low_name}")
print(f"Passed Students: {pass_students}")
print(f"Failed Students: {fail_students}")
```

```python
# Created a class named Students
class Students:
    
    # Class attribute → common for all students (shared value)
    college_name = "C.U shah college"      
    
    # Constructor → automatically runs when object is created
    def __init__(self, name, mark):   
        self.name = name      # Instance attribute → stores student name
        self.mark = mark      # Instance attribute → stores list of marks

    
    # Static method → does not use 'self' and works at class level
    @staticmethod
    def hello():
        print("Hello All")
    
    
    # Instance method → works on object data (uses self)
    def get_avg(self):              
        total = 0   # avoid using 'sum' (it is a built-in function name)
        
        for val in self.mark:
            total += val   # adding marks
        
        avg = total / len(self.mark)   # better than dividing by fixed 3
        print("Hi", self.name, "Your avg score is", avg)


# Creating objects
s1 = Students("Ankit", [95, 80, 100])  
s1.get_avg()   # calling method for s1

s2 = Students("Priya", [55, 65, 75])   
s2.get_avg()   # calling method for s2

s2.hello()     # calling static method (can also use Students.hello())


# Modifying values
s1.name = "Jayesh"   # changing name of s1

Students.college_name = "IIT Delhi"  
# modifying class attribute → affects ALL objects

s1.mark = [25, 25, 25]   # updating marks of s1
s1.get_avg()             # average will change based on new marks


# Printing values
print(s1.name, s1.mark, Students.college_name)
print(s2.name, s2.mark, Students.college_name)
print(Students.college_name)
```

```python
class Account:
    
    # Constructor → runs automatically when object is created
    def __init__(self, name, balance, account_no):
        self.name = name            # store account holder name
        self.balance = balance      # store current balance
        self.account_no = account_no  # store account number

    # Method to withdraw money
    def debit(self, amount):
        self.balance -= amount   # reduce balance
        print("Rs.", amount, "was debited from", self.name)
        print("Remaining Total", self.get_balance())  # show updated balance

    # Method to deposit money
    def credit(self, amount):
        self.balance += amount   # increase balance
        print("Rs.", amount, "was credited to", self.name)
        print("Remaining Total", self.get_balance())  # show updated balance

    # Method to check balance
    def get_balance(self):
        return self.balance   # return current balance


# Creating first account object
acc1 = Account("Ankit", 10000, 1234)

# Performing transactions
acc1.debit(1000)     # withdraw 1000
acc1.credit(500)     # deposit 500
acc1.credit(40000)   # deposit 40000
acc1.debit(10000)    # withdraw 10000


# Creating second account object
acc2 = Account("Priya", 50000, 54321)

# Performing transactions
acc2.debit(5000)     # withdraw 5000
acc2.credit(15000)   # deposit 15000
```

```python
# Example of Public Attributes in a class

class Account:
    def __init__(self, acc_no, acc_pass):
        # Public attribute: account number
        self.acc_no = acc_no
        
        # Public attribute: account password
        # ⚠️ This is not secure because it can be accessed from outside the class
        self.acc_pass = acc_pass


# Creating an object (instance) of Account class
acc1 = Account("12345", "abcd")

# Accessing public attribute (allowed)
print(acc1.acc_no)

# Accessing password directly (allowed, but NOT a good practice)
# ❌ Sensitive data like passwords should not be public
print(acc1.acc_pass)
```

```python
# Example of Private Attribute in a class

class Account:
    def __init__(self, acc_no, acc_pass):
        # Public attribute: can be accessed from outside the class
        self.acc_no = acc_no
        
        # Private attribute: cannot be accessed directly from outside
        # "__" (double underscore) makes it private (name mangling)
        self.__acc_pass = acc_pass

    # Method to access private attribute safely
    def reset_pass(self):
        # Accessing private variable inside the class (allowed)
        print(self.__acc_pass)


# Creating an object (instance) of Account class
acc1 = Account("12345", "abcd")

# Accessing public attribute (allowed)
print(acc1.acc_no)

# Accessing private attribute using method (allowed)
# Method prints password internally
print(acc1.reset_pass())  
```

### Encapsulation

```python
# Define a class named Bank
class Bank:

    # Constructor method (automatically runs when object is created)
    def __init__(self, balance):
        # Private variable __balance (Encapsulation)
        # Double underscore (__) means this variable cannot be accessed directly from outside the class
        self.__balance = balance

    # Getter method to safely access the private variable
    def get_balance(self):
        # Returns the current balance
        return self.__balance

    # Method to modify (update) the private variable
    def deposit(self, amount):
        # Adds the given amount to the balance
        self.__balance += amount


# Create an object of Bank class with initial balance 100
d = Bank(100)

# Deposit 50 into the account using method (not directly accessing balance)
d.deposit(50)

# Access the balance using getter method
print(d.get_balance())   # Output: 150


# Encapsulation Explanation:
# - __balance is hidden (private) → cannot access directly like d.__balance ❌
# - We use methods (get_balance, deposit) to access and modify it ✅
# - This protects data and controls how it is used
```

### Inheritance

```python
# Define a parent class named Animals
class Animals:
    
    # Method in parent class
    def sound(self):
        # This method is common for all animals
        print("Animal makes sounds")
        

# Define a child class named Dog which inherits from Animals
class Dog(Animals):
    
    # Method specific to Dog class
    def bark(self):
        print("Dog barks")


# Create an object of Dog class
d = Dog()

# Calling method from parent class (inherited)
d.sound()  # Dog can use this because it inherited from Animals

# Calling method from its own class
d.bark()


# Inheritance Explanation:
# - Dog class is inheriting from Animals class
# - So Dog gets access to all methods of Animals (like sound())
# - Dog can also have its own methods (like bark())
# - This helps reuse code instead of writing same code again
```

### Multiple Inheritance

```python
# Parent class 1 (Father)
class Father:
    def skills(self):
        # This method shows Father's skill
        print("Driving")


# Parent class 2 (Mother)
class Mother:
    def skills(self):
        # This method shows Mother's skill
        print("Cooking")


# Child class inheriting from both Father and Mother
class Child(Father, Mother):
    # Child does not define anything new
    # It will use methods from parent classes
    pass


# Creating object of Child class
c = Child()

# Calling skills() method using child object
# Both Father and Mother have same method name: skills()
# So Python will follow MRO (Method Resolution Order)
# It checks from LEFT to RIGHT → Father first, then Mother
# That's why Father's method will be executed
c.skills()   # Output: Driving
```

```python
# If you want both

class Father:
    def skills(self):
        print("Driving")

class Mother:
    def skills(self):
        print("Cooking")

class Child(Father, Mother):
    def skills(self):
        # Calling both parent methods manually
        Father.skills(self)   # call Father method
        Mother.skills(self)   # call Mother method


c = Child()
c.skills()
```

### Polymorphism

```python
# Define a class Dog
class Dog:
    
    # Method with same name 'sound'
    def sound(self):
        print("Bark")


# Define another class Cat
class Cat:
    
    # Method with same name 'sound'
    def sound(self):
        print("Meow")


# Create a list containing objects of different classes
animals = [Dog(), Cat()]

# Loop through each object in the list
for a in animals:
    
    # Calling the same method 'sound()' on different objects
    # Dog object will call Dog.sound() → "Bark"
    # Cat object will call Cat.sound() → "Meow"
    a.sound()


# Polymorphism Explanation:
# - Same method name 'sound()' is used in different classes
# - But behavior is different for each object
# - Dog gives "Bark", Cat gives "Meow"
# - This is called polymorphism → "one method, many forms"
```

### Getters & Setters

```python
# Getters and Setters (Encapsulation Proper)
class Account:
    def __init__(self, balance):
        self.__balance = balance  # protected (convention)

    def get_balance(self):
        return self.__balance

    def set_balance(self, new_balance):
        if new_balance >= 0:
            self.__balance = new_balance
        else:
            print("Invalid balance amount!")

acc = Account(1000)
print("Initial balance:", acc.get_balance())
acc.set_balance(1500)
print("Updated balance:", acc.get_balance())
acc.set_balance(-500)
```

### super()

```python
# Parent class
class Parent:
    def show(self):
        # This method belongs to Parent
        print("This is Parent class")


# Child class inheriting from Parent
class Child(Parent):
    def show(self):
        # super() is used to call Parent's show() method
        super().show()
        
        # This line belongs to Child
        print("This is Child method")


# Creating object of Child
c = Child()

# Calling show() method
# Step 1: Child.show() runs
# Step 2: super().show() calls Parent.show()
# Step 3: Parent prints first
# Step 4: Then Child prints
c.show()
```

```python
# Parent class
class Owner:
    def name(self):
        # This method belongs to Owner
        print("This is Owner")


# Child class inheriting from Owner
class Employee(Owner):
    def name(self):
        # Calling Owner's method using super()
        super().name()
        
        # This is Employee's own method
        print("This is Employee")


# Creating object of Employee
e = Employee()

# Calling name() method
# Step 1: Employee.name() runs
# Step 2: super().name() calls Owner.name()
# Step 3: Owner prints first
# Step 4: Then Employee prints
e.name()
```

### Error handling in Python

```python
# Basic Try-Except Example (Error Handling)

try:
    # Take input from user and convert it to integer
    a = int(input("Enter a number: "))
    
    # Calculate and print square of the number
    print("Square:", a**2)

except ValueError:
    # This block runs if user enters non-integer value (like 'abc')
    print("Please enter a valid integer.")
```

```python
# Handling Multiple Exceptions

try:
    # Take numerator input and convert it to integer
    x = int(input("Enter numerator: "))
    
    # Take denominator input and convert it to integer
    y = int(input("Enter denominator: "))
    
    # Perform division and print result
    print("Result:", x / y)

except ZeroDivisionError:
    # This block runs if user tries to divide by 0
    print("Cannot divide by zero.")

except ValueError:
    # This block runs if user enters non-numeric input (like 'abc')
    print("Enter numbers only.")
```

```python
# Using else and finally in error handling

try:
    # Take input from user and convert it to integer
    num = int(input("Enter a number: "))
    
    # Perform operation (divide by 2) and print result
    print("Half of number:", num / 2)

except ValueError:
    # This block runs if input is not a valid number (like 'abc')
    print("Invalid input.")

else:
    # This block runs only if NO error occurs in try block
    print("No error occurred.")

finally:
    # This block ALWAYS runs (whether error occurs or not)
    print("Execution completed (finally block always runs).")
```

```python
# Custom Exception Example
# Create our custom error6


# Define a custom exception by inheriting from built-in Exception class
class NegativeNumberError(Exception):
    pass   # No extra code needed, just creating a new error type

try:
    # Take input from user and convert it to integer
    n = int(input("Enter a positive number: "))
    
    # Check if number is negative
    if n < 0:
        # Manually raise (trigger) custom exception with message
        raise NegativeNumberError("Number cannot be negative.")
    
    # If number is valid, print it
    print("You entered:", n)

except NegativeNumberError as e:
    # This block handles our custom exception
    print("Custom Exception:", e)
```
