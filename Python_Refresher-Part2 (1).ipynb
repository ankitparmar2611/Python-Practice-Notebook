```python
# Slicing → list[start:end] (end not included)

colors= ["Red","Blue","Green","yellow","Orange"]

print("List :" ,colors)                                           # Prints all color list
print("colors[0:2] ->", colors[0:2])                              # Red to Blue
print("colors[:3] ->", colors[:3])                                # Red to Green
print("colors[2:] ->", colors[2:])                                # Green to Orange
print("colors[0:5] ->", colors[0:5])                              # Red to Orange
print("colors[-4:-1] ->", colors[-4:-1])                          # Yellow to Blue in reverse order
print("colors[-3:-1] ->", colors[-3:-1])                          # Green to Yellow in reverse order
print("Fetching Every Second Element", colors[::2])               # Fetching every second element
print("colors Green:", colors[2])                                 # Fetch only Green color
```

```python
# Union, intersection, difference, symmetric_difference

A={1,2,3,4}
B={3,4,5,6}

print("A union B: ", A|B)                           # Concatinate both varaible  A and B
print("A intersection B :", A & B)                  # Common elements between A and B
print("A difference B :", A - B)                    # Elements which are present in A but not there in B
print("B difference A :", B - A)                    # Eelements which are present in B but not there in A
print("Symmetric difference of B", A ^ B)           # Elements other then common elements present in A and B.
print("Symmetric difference of B", B ^ A)           # Elements other then common elements present in A and B.
```

```python
# Subset, superset and disjoint
A = {1, 2, 3}
B = {1, 2, 3, 4, 5}

print("A ⊆ B:", A.issubset(B))                        # Check if all elements of A are inside B if yes then its True
print("B ⊇ A:", B.issuperset(A))                      # Check if B contains all elements of A if yes then its True
print("A and B disjoint?", A.isdisjoint({6, 7, 8}))   # No common elements between A={1,2,3} and {6,7,8}, so its True
```

```python
# 'continue' skips the current iteration (Times)

# Odd numbers example

for i in range(1,8):                  # Generates numbers from 1 to 7 (8 excluded)
    if i % 2 == 0:                    # Checks If number is even (divisible by 2)
        continue                      # Skips the even numbers and continue
    print("Odd numbers are", i)




# Even numbers example

for i in range(1,8):
    if i % 2 != 0:                    # checks if the number is odd
        continue                      # Skips the odd numbers and continue
    print("Even numbers are :", i)    
```

```python
# Searching for an item using loop + else

# Example 1

# List of numbers
numbers = [10, 20, 30, 40, 50]

# Number we want to search in the list
target = 30

# Loop through each number in the list
for num in numbers:

    # Check if current number equals the target
    if num == target:
        print(f"{target} found in list")   # Print if found
        break                              # Stop the loop immediately

# This else belongs to the for loop
# It runs only if the loop finishes without hitting break
else:
    print(f"{target} not found")

# Example 2

names = ["ankit","ajay","karan","jaydip"]
target = "ankit"

for name in names:
    if name == target:
       print(f"{target} found in the list")
       break
else:
    print(f"{target} not found in list")

# Example 3
tuples = {"set","tuple","dictionary","list"}
target = "tuple"

for tuple in tuples:
    if tuple == target:
        print(f"{target} is found")
        break
else:
    print(f"{tuple} is not found")
```

```python
# Student Result Analyser

# 1️⃣ Count how many students passed (marks ≥ 50)
# 2️⃣ Count how many students failed (marks < 50)
# 3️⃣ Find the highest mark
# 4️⃣ Calculate the average marks
marks = [78, 45, 90, 33, 67, 82, 49]
passed = 0
failed = 0
highest = marks[0]
average = 0
count = 0

for mark in marks:
    count = count + 1
    if mark > highest:
        highest = mark
    if mark >= 50:
        passed += 1
    if mark < 50:
        failed += 1
    average = average + mark
    result = average / count
    
print(passed)
print(failed)
print(highest)
print(result)


```

```python
# Write a program to:

# 1️⃣ Count orders above 1000
# 2️⃣ Find lowest sale
# 3️⃣ Calculate total revenue
# 4️⃣ Count orders below 500

sales = [1200, 450, 800, 1500, 300, 950]

order_above_1000 = 0
lowest_sale = sales[0]
total_rev = 0
order_below_500 = 0

for order in sales:
    total_rev = total_rev + order
    if order < lowest_sale:
        lowest_sale = order
    if order > 1000:
        order_above_1000 += 1
    if order < 500:
        order_below_500 += 1
        
print(f"Order above 1000 : {order_above_1000}")        
print(f"Lowest sale : {lowest_sale}")
print(f"Total Revenue : {total_rev}")
print(f"Orders below 500 : {order_below_500}")
```

```python
# Count numbers greater then 50.
numbers = [23, 67, 45, 89, 12, 90]

i = 0
total = 0

while i < len(numbers):
    if numbers[i] > 50:
        total += 1              # This is used to count numbers which are greater then 50
    i += 1
print(total)    
```

```python
# Calculate sum of list
numbers = [10, 20, 30, 40]

i = 0
total = 0

while i < len(numbers):
    total = total + numbers[i]   # This is used to do addition of all numbers
    i += 1
print(total)
```

```python
# Count Odd number
numbers = [11, 24, 37, 40, 55]

i = 0
count = 0

while i < len(numbers):
    if numbers[i] % 2 != 0:
        count += 1             # Used to count total odd numbers
    i += 1
print(count)
```

```python
# Find if the number exist

numbers = [4, 8, 15, 16, 23]

i = 0
found = False

while i < len(numbers):
    if numbers[i] == 7:
        found = True
        break
    i += 1
print(found)    
```

```python
# Remove duplicates from list
numbers = [1, 2, 2, 3, 4, 4, 5]

i = 0
new_list = []

while i < len(numbers):
    if numbers[i] not in new_list:  # It will check that if numbers[i] exits in new_list.
        new_list.append(numbers[i]) # It will creat new list and append the unique numbers
    i += 1
print(new_list)    


```

```python
# Keys,values,items

marks = {
    "Ankit": 85,
    "Priya": 90,
    "Rahul": 78
}

for k in marks.keys():
    print(f"Keys : {k}")
for v in marks.values():
    print(f"Values :{v}")

for name,score in marks.items():
    print(f"Name Score : {name}:{score}")
```

```python
# Total marks and highest marks
marks = {
    "Ankit": 85,
    "Priya": 90,
    "Rahul": 78
}

total = 0

highest = 0

for score in marks.values():
    total += score
print(f"Total marks :{total}")

for score in marks.values():
    if score > highest:
        highest = score
print(f"Highest marks :{highest}")        
    
```

```python
# List containing fruit names (some fruits appear multiple times)
fruits = ["apple", "banana", "apple", "mango", "banana", "apple"]

# Empty dictionary to store fruit counts
count = {}

# Loop through each fruit in the list
for fruit in fruits:
    
    # Check if the fruit already exists as a key in the dictionary
    if fruit in count:
        # If it exists, increase its count by 1
        count[fruit] += 1
    else:
        # If it does not exist, add it to the dictionary with count = 1
        count[fruit] = 1

# Print the final dictionary with fruit counts
print(count)
```

```python
# List containing programming language names (some appear multiple times)
words = ["python","java","python","c","java","python"]

# Empty dictionary to store how many times each word appears
count = {}

# Loop through each word in the list
for word in words:

    # Check if the word already exists as a key in the dictionary
    if word in count:
        # If it exists, increase its count by 1
        count[word] += 1
    else:
        # If it does not exist, add the word to the dictionary with value 1
        count[word] = 1

# Print the final dictionary showing word counts
print(count)
```

```python
# Print Even and Odd numbers separatly
numbers = [1,2,3,4,5,6,7,8]

even = []
odd = []

for num in numbers:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)
print(f"Even numbers : {even}")
print(f"Odd numbers : {odd}")

```

```python
# Instead of two lists, we can store them inside a dictionary.
numbers = [1,2,3,4,5,6,7,8]

result = {
    "even": [],
    "odd": []
}

for num in numbers:
    if num % 2 == 0:
        result["even"].append(num)
    else:
        result["odd"].append(num)

print(result) 
```

```python
# Numbers greater the 10 and numbers Less the equal to 10
numbers = [12,5,8,15,20,3,10]
result = {
    "Greater_then_10": [],
    "Less_then_equal_to_10": []
}

for num in numbers:
    if num > 10:
        result["Greater_then_10"].append(num)
    else:
        result["Less_then_equal_to_10"].append(num)
print(result)        
```

```python
# Find the student with highest marks
students = [
{"name":"Ankit","marks":85},
{"name":"Priya","marks":40},
{"name":"Rahul","marks":78},
{"name":"Komal","marks":90},
{"name":"Karan","marks":99},    
]

highest = 0
top_student = ""

for student in students:
    name = student["name"]
    score = student["marks"]
    if score > highest:
        highest = score
        top_student = name   
print(f"Top student : {top_student} \nMarks : {score}")
    
```

```python
# Find total marks # Find highest marks 
# Find the lowest marks 
# Find highest grade with name and marks

students = [
    {"name": "Ankit","marks": 99,"Grade": "A"},
    {"name": "Karan","marks": 50,"Grade": "C"},
    {"name": "Kishan","marks": 34,"Grade": "D"},
    {"name": "Priya","marks": 80,"Grade": "B"},
    {"name": "Komal","marks": 85,"Grade": "B"},
    {"name": "Jayesh","marks": 70,"Grade": "B"},
    {"name": "Jagu","marks": 30,"Grade": "C"}
]

total_marks = 0
highest_marks = 0
lowest_marks = students[0]["marks"]
target_grade_A = "A"
target_grade_B = "B"
target_grade_C = "C"
target_grade_D = "D"
count_grade_A = 0
count_grade_B = 0
count_grade_C = 0
count_grade_D = 0
average_marks = 0
student_marks = {
"Student_high_marks": [],
"Student_low_marks": []
}
students_list = {
    "A_grade_list" : [],
    "B_grade_list" : [],
    "C_grade_list" : [],
    "D_grade_list" : []
    
}

for student in students:
    name = student["name"]
    mark = student["marks"]
    grade = student["Grade"]
    
    user_detail = name,mark,grade
    
    total_marks += mark
    average_marks = total_marks / len(students)
    
    if mark > highest_marks:
        highest_marks = mark
        student_marks["Student_high_marks"] = [user_detail]
        
    if mark < lowest_marks:
        lowest_marks = mark
        student_marks["Student_low_marks"]= [user_detail]
    if grade == target_grade_A:
        count_grade_A += 1
        students_list["A_grade_list"].append(name)

    if grade == target_grade_B:
        count_grade_B += 1
        students_list["B_grade_list"].append(name)
    if grade == target_grade_C:
        count_grade_C += 1
        students_list["C_grade_list"].append(name)
    if grade == target_grade_D:
        count_grade_D += 1    
        students_list["D_grade_list"].append(name)
         
                   
print(f"Total marks : {total_marks}")
print(f"Highest marks : {highest_marks}")
print(f"Lowest marks : {lowest_marks}")
print(f"Total A Grade : {count_grade_A}")
print(f"Total B Grade : {count_grade_B}")
print(f"Total C Grade : {count_grade_C}")
print(f"Total D Grade : {count_grade_D}")
print(f"Average marks : {average_marks}")
print(f"A Grade list students : {students_list["A_grade_list"]}")
print(f"B Grade list students : {students_list["B_grade_list"]}")
print(f"C Grade list students : {students_list["C_grade_list"]}")
print(f"D Grade list students : {students_list["D_grade_list"]}")
print(f"Student with highest mark {student_marks["Student_high_marks"]}")
print(f"Student with low mark {student_marks["Student_low_marks"]}")
```
